# Visualisation des paires de caracteristiques

```python
import seaborn as sns
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data_file_path = '../data/raw/dataTp2.csv'
if os.path.exists(data_file_path):
    print("File exists : ", data_file_path)
```

```python
# Chargement des données
data = pd.read_csv(data_file_path)
data.info
```

```python
# Liste des caractéristiques
features = ["valence_intensity", "fear_intensity", "anger_intensity", "happiness_intensity", "sadness_intensity"]

# Création des paires uniques
pairs = [(features[i], features[j]) for i in range(len(features)) for j in range(i+1, len(features))]

# Visualisation de chaque paire
for feature_x, feature_y in pairs:
    plt.figure(figsize=(6, 6))
    sns.jointplot(data=data, x=feature_x, y=feature_y, kind="hex", cmap="viridis")
    plt.suptitle(f"Relation entre {feature_x} et {feature_y}", y=1.05)
    plt.show()
```

# Visualisation superposee

```python

# Visualisation des distributions superposées pour chaque paire
for feature_x, feature_y in pairs:
    # Calcul du coefficient de corrélation
    correlation = np.corrcoef(data[feature_x], data[feature_y])[0, 1]
    
    # Création de la figure
    plt.figure(figsize=(8, 6))
    sns.histplot(data[feature_x], color="green", label=feature_x, kde=True, binwidth=0.05)
    sns.histplot(data[feature_y], color="blue", label=feature_y, kde=True, binwidth=0.05)
    
    # Ajout du titre et des labels
    plt.title(f"Distribution conjointe de {feature_x} et {feature_y}\nCoefficient de corrélation: {correlation:.2f}")
    plt.xlabel("Intensity")
    plt.ylabel("Frequency")
    plt.legend()
    
    # Affichage
    plt.show()
```
